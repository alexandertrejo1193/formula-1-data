### Carreras

```python
races1={'year':[],'round':[],'raceName':[],'circuitId':[],'date':[],'time':[],'dateq':[],'timeq':[]}
for year in list(range(2023,2024)):
    url='http://ergast.com/api/f1/{}.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        races1['year'].append(int(item['season']))
        races1['round'].append(int(item['round']))
        races1['raceName'].append(item['raceName'])
        races1['circuitId'].append(item['Circuit']['circuitId'])
        races1['date'].append(item['date'])
        races1['time'].append(item['time'])
        races1['dateq'].append(item['Qualifying']['date'])
        races1['timeq'].append(item['Qualifying']['time'])
racess1=pd.DataFrame(races1)
racess1['date'] = racess1['date'].astype('datetime64[ns]')
racess1['dateq'] = racess1['dateq'].astype('datetime64[ns]')
racess1.to_csv('races_2023_contiempos.csv',index=False)
racess1
```
