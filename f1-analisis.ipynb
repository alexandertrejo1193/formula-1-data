```python
races={'year':[],'round':[],'raceName':[],'circuitId':[],'lat':[],'long':[],'country':[],'date':[],'url':[]}
for year in list(range(year1,year2)):
    url='http://ergast.com/api/f1/{}.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        races['year'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(float(item['Circuit']['Location']['lat']))
        races['long'].append(float(item['Circuit']['Location']['long']))
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
racess=pd.DataFrame(races)
racess.head()
racess['date'] = racess['date'].astype('datetime64[ns]')
racess.head() 
```

### Temporadas

```python
warnings.filterwarnings("ignore")
url='http://ergast.com/api/f1/seasons.json?limit=1000'
response = requests.get(url)
content=json.loads(response.content)
Season ={'year':[],'url':[]}
for item in content['MRData']['SeasonTable']['Seasons']:
    Season['year'].append(int(item['season']))
    Season['url'].append(item['url'])
Seasons=pd.DataFrame(Season)
Seasons.head()
```

### Conductores

```python
drivers={'year':[],"DriverId":[],"Number":[],"code":[],"forename":[],"surname":[],"dob":[],"nationality":[],"url":[]}
for year in list(range(year1,year2)):
    url='http://ergast.com/api/f1/{}/drivers.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['DriverTable']
    for i in range(0,len(path["Drivers"])):
            drivers["year"].append(int(path['season']))
            try:
                drivers["Number"].append(int(path['Drivers'][i]["permanentNumber"]))
            except:
                drivers["Number"].append("")
            drivers["DriverId"].append(path['Drivers'][i]["driverId"])
            try:
                drivers["code"].append(path['Drivers'][i]["code"])
            except:
                drivers["code"].append("")
            drivers["forename"].append(path['Drivers'][i]["givenName"])
            drivers["surname"].append(path['Drivers'][i]["familyName"])                       
            drivers["dob"].append(path['Drivers'][i]["dateOfBirth"])                        
            drivers["nationality"].append(path['Drivers'][i]["nationality"])
            drivers["url"].append(path['Drivers'][i]["url"]) 
driverss=pd.DataFrame(drivers)
driverss['dob'] = driverss['dob'].astype('datetime64[ns]')
driverss.head() 
```

### Constructores

```python
constructors={"year":[],"constuctorId":[],"name":[],"nationality":[],"url":[],}
warnings.filterwarnings("ignore")
for year in list(range(year1,year2)):
    url='http://ergast.com/api/f1/{}/constructors.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['ConstructorTable']
    for item in range(0,len(path['Constructors'])):
        constructors["year"].append(int(path['season']))
        constructors["constuctorId"].append(path['Constructors'][item]['constructorId'])
        constructors['name'].append(path['Constructors'][item]['name'])
        constructors['nationality'].append(path['Constructors'][item]['nationality'])
        constructors['url'].append(path['Constructors'][item]['url'])
    
constructorss=pd.DataFrame(constructors)
constructorss.head()
```

### Circuitos

```python
circuits={'year':[],'circuitId':[], #'circuitRef':[],
           'name':[],'location':[], 'country':[],
           'lat':[], 'lng':[], 'alt':[],
           'url':[]}
warnings.filterwarnings("ignore")
for year in range(year1,year2):
    url= 'http://ergast.com/api/f1/{}/circuits.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    path=content['MRData']['CircuitTable']
    for item in range(0,len(path['Circuits'])):
        circuits['year'].append(int(path['season']))
        circuits['circuitId'].append(path['Circuits'][item]['circuitId'])
        circuits['name'].append(path['Circuits'][item]['circuitName'])
        circuits['location'].append(path['Circuits'][item]['Location']['locality'])
        circuits['country'].append(path['Circuits'][item]['Location']['country'])
        circuits['lat'].append(float(path['Circuits'][item]['Location']['lat']))
        circuits['lng'].append(float(path['Circuits'][item]['Location']['long']))
        circuits['alt'].append(path['Circuits'][item]['Location']['locality'])
        circuits['url'].append(path['Circuits'][item]['url'])
    
circuitss=pd.DataFrame(circuits)
circuitss.head()
```

```python
import warnings
qualifiying={"year":[],"driverId":[],"ConstructorId":[],"number":[],"position":[],"qualifyId":[],"raceId":[],"q1":[],"q2":[],"q3":[]}
warnings.filterwarnings("ignore")
for year in range(year1,year2):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    lenpath=len(content['MRData']['RaceTable']['Races'])
    links=[]
    for i in range(0,lenpath):
        url2=url3+'/{}/qualifying.json?limit=100'
        links.append(url2.format(i))
    for j in links:
        response = requests.get(j)
        content=json.loads(response.content)
        for item in content['MRData']['RaceTable']['Races']:
            for seconditem in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
                qualifiying["year"].append(int(item['season']))
                qualifiying["qualifyId"].append(int( item['round']))
                qualifiying["raceId"].append(item['raceName']) 
                qualifiying["driverId"].append(seconditem['Driver']['driverId'])   
                qualifiying["ConstructorId"].append(seconditem['Constructor']['constructorId'])
                qualifiying["number"].append(int(seconditem['number']))
                qualifiying["position"].append(int(seconditem['position']))
                qualifiying["q1"].append(seconditem['Q1'])
                try:
                    qualifiying["q2"].append(seconditem['Q2'])
                except:
                    qualifiying["q2"].append("")
                try:
                    qualifiying["q3"].append(seconditem['Q3'])
                except:
                    qualifiying["q3"].append("")

qualifiyingg=pd.DataFrame.from_dict(qualifiying,orient='index')
qualifiyingg=qualifiyingg.transpose()
qualifiyingg.head()
```
